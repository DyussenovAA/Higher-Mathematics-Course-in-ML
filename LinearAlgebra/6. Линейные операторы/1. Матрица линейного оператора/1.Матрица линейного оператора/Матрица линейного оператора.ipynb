# Линейное отображение и его образ

Пусит $V$ и $W$ - Линейные пространства над одним и тем же полем $F$. 

Отображение $$T:V \to W$$ 

называется линейным, если для любых $x, y \in V$ и любых $\alpha, \beta \in F$ выполняется:

$$T(\alpha x + \beta y) = \alpha T (x) + \beta T (y)$$

Линейное отображение сохраняет линейные комбинации. Если вектор собран из других векторов, его образ собирается из их образов теми же коэффициентами.

Следствия:

- Если $T$ линейно, то:

$$T(0) = 0$$
$$T(0) = T(0 \cdot x) = 0 \cdot T(x)= 0$$

- $T(-x) = -T(X)$

Образ линейной комбинации полностью определяется образами базиса 

Если $\{e_1, ..., e_n \}$ - базис $V$, то любой $x$ представим как 

$$x = \sum_j x_j e_j$$

тогда 

$$T(x) = \sum_j x_j T(e_j)$$

Линейное отображение полностью определяется значениями на базисе. 

---

Образ Линейного отображения

Образ отображения $T$ - это множество 

$$Im  T = \{T(x) | x \in V \} \subseteq W$$

множество всех векторов в конечном пространстве $W$, в которые мы «попали», применив оператор $T$ к векторам из исходного пространства $V$.

Т.е. векторы которые можно получить как результат действия $T$ 

Линейное отображение — это функция между линейными пространствами, сохраняющая линейные комбинации.

Образ — это множество всех возможных значений этой функции.

Образ всегда является подпространством.

Размерность образа = число независимых направлений, которые не уничтожаются отображением.

# П1. Матрица линейного оператора

**Опеределение 1.** Линейным оператором (линейным преобразованием) в векторном пространстве $V$ называется линейное отображение пространства $V$ в себя.

Т.е. линейный оператор - это отображение $\mathcal{A}: V \to V$, удовлетворяющее условиям:
1) $\mathcal{A} (x + y) = \mathcal{A} x + \mathcal{A} y$ для любых $x, y \in V$;
2) $\mathcal{A} (\lambda x) = \lambda \mathcal{A} x$ для любых $x \in V$, $\lambda \in K$

Если в пространстве $V$ выбран базис $\{e_1, ..., e_2\}$, то линейный оператор может быть задан матрицей. 

1. Главная суть, в том, что линейный оператор это правило изменения состояния, которое сохраняет сложение и масштабирование. Это проявляется в том, что система предсказуема относительно суперпозиции 

Сигнал A - эффект $X$

Сигнал $B$ - эффект $Y$

$$A + B \to X + Y$$

или 

$$\mathcal{A}(\alpha x + \beta y) = \alpha \mathcal{A}x + \beta \mathcal{A} y$$

2. Отображение в себя означает, Что пространство не меняется, а только координаты точек меняются. Геометрически это: растяжение, поворот, отражение, проекция, сжатие, но не меняется размерность, кроме случаев когда размерность может "захлопнуться". 

3. Если выбрать базис, оператор можно задать матрицей и любой линейный оператор полностью определяется тем, куда он отправляет базисные векторы.

**Определение 2.** Матрицей линейного оператора $\mathcal{A}$ в базисе $\{e_1, ..., e_n\}$ называется матрица $A = (a_{ij})$, определяемая из равенств $$\mathcal{A} e_j = \sum_{i} a_{ij} e_i$$
Иначе говоря, в j-м столбце матрицы $A$ стоят координаты вектора $\mathcal{A} e_j$ в базисе $\{e_1, ..., e_n\}$. Здесь фигурирует только один базис. 

В матричной форме: $$(\mathcal{A} e_1,..., \mathcal{A} e_n) = (e_1, ..., e_n)A$$

В левой части - список образов базиса, в правой - базис умноженный на матрицу. 

Для любых векторов $f_1, ..., f_n \in V$ существует единственный линейный оператор $\mathcal{A}$ переводящий базисные векторы $e_1, ..., e_n$ в $f_1,... ,f_n$ соответственно. Э то оператор, переводящий каждый вектор $x = \sum_i x_i e_i$ в вектор $\sum_i x_i f_i$. Следовательно, линейный оператор однозначно определяется своей матрицей, и любая квадратная матрциа порядка $n$ является матрицей некоторого линейного оператора (в данном базисе).

Найдем явное выражение координат образа $y = \mathcal{A} x$ вектора $x$. При $x = \sum_j x_j e_j$ имеем $$y = \sum_j x_j \mathcal{A} e_j = \sum_{i, j} a_{ij} x_j e_i = \sum_i y_i e_i$$

где $$y_i = \sum_j a_{ij} x_j$$

Если обозначить через $X$ и $Y$ столбцы координат векторов $x$ и $y$ соответственно, то равенства можно переписать в следующей матричной форме:
$$Y = AX$$

Где $A - $ матрица перехода.
