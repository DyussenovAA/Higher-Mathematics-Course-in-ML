#### **1. Определение, геометрическая интерпретация комплексных чисел в алгебраической форме.Сложение, вычитание, умножение, деление, возведение в степень комплексных чисел в алгебраической форме.**

**1.1 Определение:** Комплексное число - выражение вида $a + bi$, где $a, b$ - вещественные числа $(a, b \in \R)$, при этом число $a$ называется вещественной частью числа и Обозначается как $Re \ z$, а число $b$ называется мнимой частью числа $z$ и обозначается как $Im \ z$.<br>
$a + bi$ - называется алгебраической формой.

Допустим, что <br>

$$z_1 = a_1 + b_1 \cdot i, \ z_2 = a_2 + b_2 \cdot i$$

тогда $$z_1 = z_2 \;\Leftrightarrow\; \left\{\begin{array}{l} a_1 = a_2 \\ b_1 = b_2 \end{array}\right.$$

комплексные числа равны, когда равны их вещественные и мнимые части

**1.2. Геометрическая интерпретация:**

Комплексные числа можно изобразить точками или векторами на плоскости. А именно, число $z = a + bi$ изображается точкой или вектором с декртовыми координатами $(a, b)$. При этом ось абсцисс соответствует множеству вещественных чисел, ордината точки равна мнимой части соответствующего числа.

сложение комплексных чисел соответствует сложению векторов по правилу параллелограмма и треугольника и сложению их соответствующих

**1.3. Сложение, вычитание, умножение, деление, возведение в степень комплексных чисел в алгеббраической форме.**

Сложение и вычитание:

$z_1 = a_1 + b_1 i, \ z_2 = a_2 + b_2$

тогда:

$$z_1 + z_2 = (a_1 + a_2) + (b_1 + b_2)i, $$
$$z_1 - z_2 = (a_1 - a_2) + (b_1 - b_2)i $$

Умножение комплесных чисел: 

$$i^2 = -1$$
в общем формула выглядит вот так: 
$$(a + bi) \cdot (c + di) = (ac - bd) + (ad + bc)i$$

Сопряжение комплексных чисел:

Сопряженным к комплексному числу $a+bi$ называют число $a - bi$.
В общем виде комплексное число, сопряженное к числу $z$, обозначается $\hat{z}$

Деление комплесных чисел:

$$\frac{z_1}{z_2} = \frac{a_1 + b_1 i}{a_2 + b_2 i} = \frac{(a_1 + b_1 i)(a_2 - b_2 i)}{a_2^2 + b_2^2}$$


$$\frac{z_1}{z_2} = \frac{a_1 a_2 + b_1 b_2}{a_2^2 + b_2^2} + \frac{b_1 a_2 - a_1 b_2}{a_2^2 + b_2^2} i$$

Степень:

$$z^n = z \cdot z \cdot z \ ... \ n$$

#### **3. Модуль и аргумент комплексного числа. Тригонометрическая форма комплексного числа. Умножение и возведение в степень комплексных чисел в тригонометрической форме.**

Модулем комплесного числа $z = a + bi$ называется длина вектора, изображающего это число. Модуль числа $z$ обозначается $|z|$ и оно равно:
$$|c| = \sqrt{a^2 + b^2}$$

Аргументом комплексного числа называется угол, образуемый соответствующим вектором с положительным направленимем оси абсцисс. Аргумент определен с точностью до прибавления целого кратного $2 \pi$. Аргумент числа 0 не определен. Аргумент числа $z$ обозначается через $arg \ z$ <br>
Пусть $r$ и $\phi$ - модуль и аргумент числа $z$, тогда $$a = r \ \cos \phi, \ \ b = r \ \sin \phi$$

откуда 

$$z = r(\cos \phi + i \sin \phi)$$

Такое представление называется тригонометрической формой. Так как тригонометрическая форма данного комплексного числа определена однозначно с точностью до прибавления к $\phi$ целого кратного 2 $\pi$, то при $r_1, r_2 > 0$:
$$r_1 (\cos \phi_1 + i \sin \phi_1)=r_2(\cos \ phi_2 + i \sin \phi_2) \Longleftrightarrow \{r_1 = r_2, \phi_1 = \phi_2 + 2\pi k , k \in \Z \}$$

Умножение комплексных чисел: 

$$r_1(\cos \phi_1 + i \sin \phi_1) \cdot r_2(\cos \phi_2 + i \sin \phi_2) = r_1 r_2 (\cos (\phi_1 + \cos \phi2) + i \sin (\phi_1 + \phi_2))$$
