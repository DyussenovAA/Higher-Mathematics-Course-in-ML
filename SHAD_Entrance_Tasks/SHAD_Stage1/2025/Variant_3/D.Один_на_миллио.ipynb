# Один на миллион

Найти раидус сходимости ряда $\sum_{n=1}^{\infty} x^{[n \ln n]} \frac{(n!)^2}{(3n)!}$

Введенное вами число должно отличаться от правильного ответа не более чем на $10^{-9}$

Прологрифмируем: $[n ln(n)] ln(x) + 2ln(n!) - ln((3n)!)$

Эквивалетность: $~ n ln(n) ln(x) + 2 ln(n!) - ln((3n)!)$
