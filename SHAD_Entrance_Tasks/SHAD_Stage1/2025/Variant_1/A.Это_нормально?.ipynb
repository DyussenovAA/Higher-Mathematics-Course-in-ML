### [Структура $G(x)$](#my-custom-id)
С точки зрения математического анализа, функция $G(x)$ представляет собой интегральное преобразование типа свертки с переменными пределами интегрирования. Общий вид данной конструкции:

$$G(x)=C\int _{a(x)}^{b(x)}F(t)k(x,t)dt$$

1. Оператор сглаживания
В контексте теории сигналов и статистики данное выражение интерпретируется как процесс сглаживания исходного распределения $F(t)$ при помощи весового ядра $k(x,t)$. Интеграл "усредняет" значения $F(t)$ в окрестности точки $x$, что приводит к потере локальной резкости (дифференцируемости), но делает итоговую функцию $G(x)$ более гладкой.

2. Свойства пределов и окна наблюдения

Пределы интегрирования $a(x)=x-1$ и $b(x)=x+1$ формируют так называемое "скользящее окно"  постоянной ширины $L=2$.

Инвариантность к сдвигу: Тот факт, что границы линейно зависят от $x$ с единичным коэффициентом, указывает на то, что форма окна не деформируется при изменении $x$, а лишь перемещается вдоль оси абсцисс.

Финитность: Окно ограничено, что означает, что в каждой конкретной точке $x$ значение функции распределения $G(x)$ определяется исключительно локальным поведением $F(t)$ в радиусе $\pm 1$.

3. Вероятностная интерпретация через свертку

Структура $G(x)$ полностью эквивалентна функции распределения суммы двух независимых случайных величин $X=Y+Z$, где:

- $Y$ - случайная величина с функцией распределения $F(y)$.

- $Z$ - случайная величина (шум), имеющая плотность $f_{Z}(z)=Ck(z)$ на интервале $[-1,1]$.С математической точки зрения это классическая свертка функции распределения и плотности: $F_{X}(x)=(F_{Y}*f_{Z})(x)$.

### [Анализ заданной функции $G(x)$](#my-custom-id1)

Рассмотрим компоненты конкретного выражения:

$$G(x)=C\int _{x-1}^{x+1}F(t)\>\mathrm{arctg}\>|x-t|dt$$

1. Подынтегральная функция $F(t)$

Функция $F(t)=1-e^{-t^{2}/2}$ при $t\ge 0$ является интегральной функцией распределения Рэлея с параметром масштаба $\sigma =1$.

- Она характеризуется наличием положительной асимметрии и экспоненциально убывающим "хвостом".

- Все моменты этой величины существуют, что гарантирует существование дисперсии $D[Y]=2-\pi /2$.

2. Ядро арктангенса и роль модуля

Ядро системы задано как $k(x,t)=\>\mathrm{arctg}\>|x-t|$. Проанализируем его свойства:

- Четность: Исходная функция $\mathrm{arctg}\>(u)$ является нечетной. Однако введение модуля $|u|$ делает ядро четной функцией относительно разности $(x-t)$.

- Симметрия шума: В вероятностном смысле четность ядра означает, что случайная величина $Z$ распределена симметрично относительно нуля. Это критически важное свойство, из которого следует, что математическое ожидание $E[Z]=0$.

- Форма распределения: График $\mathrm{arctg}\>|u|$ на интервале $[-1,1]$ имеет V-образную форму. Это говорит о том, что крайние значения в "окне сглаживания" имеют больший статистический вес, чем центральные.

3. Роль нормировочной константы $C$

Константа $C$ не просто масштабирует функцию, она выполняет роль статистической нормировки. Для того чтобы $G(x)$ была корректной функцией распределения (то есть $\lim _{x\rightarrow \infty }G(x)=1$), необходимо, чтобы "вес" ядра был нормирован к единице:

$$C\int _{-1}^{1}\>\mathrm{arctg}\>|z|dz=1$$

Это условие превращает ядро в полноценную плотность вероятности $f_{Z}(z)$.

Итог анализа

Благодаря структуре свертки, задача нахождения дисперсии $D[G]$ сводится к суммированию дисперсий двух независимых компонент:$D[G]=D[\text{Rayleigh}]+\frac{\int _{-1}^{1}z^{2}\>\mathrm{arctg}\>|z|dz}{\int _{-1}^{1}\>\mathrm{arctg}\>|z|dz}$


Задания такого типа на олимпиадах/экзаменах часто строятся так, чтобы проверять понимание свойств свёртки и дисперсии, нам не нужно брать громоздкие интегралы, вычислять первый, второй момент. Это гораздо сложнее, и такие задачи обычно не дают без возможности упрощения. Это "Brute force" подход, который математически верен, но нерационален по времени и сложности.

### [Интеграл свертки](#my-custom-id2) 
В теории вероятностей свертка двух функций описывает закон распределения суммы двух независимых случайных величин $X=Y+Z$. Если $F_{Y}(y)$ - функция распределения величины $Y$, а $f_{Z}(z)$ - плотность распределения величины $Z$, то функция распределения их суммы $X$ определяется как:

$$F_{X}(x)=\int _{-\infty }^{+\infty }F_{Y}(x-z)f_{Z}(z)dz$$

1. Дуализм плотности и распределения 

Важно различать два вида свертки, используемых в задаче: 

- Свертка плотностей: Если известны плотности $f_{Y}$ и $f_{Z}$, то плотность суммы $f_{X}$ есть 

$$f_{X}(x)=\int _{-\infty }^{+\infty }f_{Y}(x-t)f_{Z}(t)dt$$

- Свертка функции распределения и плотности: Как в нашем случае, когда одна величина задана через распределение $F(t)$, а "окно" сглаживания - через плотность ядра. Это позволяет сразу получить итоговую функцию распределения $G(x)$. 

2. Коммутативность и симметрия

Операция свертки обладает свойством коммутативности: $(f*g)=(g*f)$. Это означает, что не имеет значения, какую величину мы считаем "основным сигналом", а какую - "сглаживающим шумом". В нашей задаче это позволяет интерпретировать $F(t)$ как распределение $Y$, а арктангенс - как плотность $Z$. 

3. Роль пределов интегрирования в свертке В классической формуле свертки интегрирование ведется по всей прямой. Однако в заданном выражении $G(x)$ пределы ограничены интервалом $[x-1,x+1]$. Это указывает на то, что плотность второй случайной величины $Z$ является финитной, то есть:

$$f_{Z}(z)=\begin{cases}C\cdot \>\mathrm{arctg}\>|z|,&z\in [-1,1]\ 0,&z\notin [-1,1]\end{cases}$$

Таким образом, конечные пределы интеграла - это не просто границы области, а прямое следствие ограниченности носителя (support) плотности "шума". 

4. Алгебраическое преимущество 

Идентификация интеграла как свертки дает доступ к свойству аддитивности моментов. Для независимых величин: 

- Математическое ожидание суммы равно сумме ожиданий: $E[X]=E[Y]+E[Z]$.

- Дисперсия суммы равна сумме дисперсий: $D[X]=D[Y]+D[Z]$. 

Это свойство является критическим для решения, так как оно переводит задачу из области интегрального исчисления в область алгебраического суммирования числовых характеристик распределений.

Проверка условия корректности свертки в нашей задачи:
- Ядро $\arctg |x - t|$ зависит только от $|x - t|$ - инвариантность относительно сдвига.
- Пределы $x \pm 1$ после замены $u = x - t$ стали постоянными $[-1, 1]$.
- $h(u) = \arctg |u|$ неотрицательна на $[-1, 1]$ - можно нормировать в плотность.
- $F(t)$ - действительно функция распределения (неубывающая, непрерывная справа, пределы $0$ и $1$).

### Вероятностная интерпретация и декомпозиция $G(x)$

На основе проведенного анализа структуры и свойств свертки мы выдвигаем гипотезу: случайная величина $X$, имеющая функцию распределения $G(x)$, представима в виде суммы двух независимых случайных величин:

$$X=Y+Z$$

1. Идентификация слагаемых 

Мы сопоставляем компоненты заданного интеграла $G(x)=C\int _{x-1}^{x+1}F(t)\cdot \>\mathrm{arctg}\>|x-t|\,dt$ со стандартной формулой распределения суммы $F_{X}(x)=\int _{-\infty }^{+\infty }F_{Y}(t)f_{Z}(x-t)dt$: 

- Величина $Y$ (Сигнал): Её функция распределения задана непосредственно под интегралом. Это $F_{Y}(t)=F(t)$.

    - Согласно условию: $F_{Y}(t)=1-e^{-t^{2}/2}$ при $t\ge 0$.
    
    - Это распределение Рэлея, которое отвечает за "основное положение" случайной величины.
    
- Величина $Z$ (Аддитивный шум): Её плотность вероятности извлекается из ядра интеграла.

    - Плотность $f_{Z}(u)$ пропорциональна функции $\mathrm{arctg}\>|u|$.
    
    - Важнейший аспект - носитель плотности. Поскольку интеграл берется в пределах от $x-1$ до $x+1$, плотность $f_{Z}$ отлична от нуля только на интервале $[-1,1]$.
    
    - Таким образом: $f_{Z}(z)=C\cdot \>\mathrm{arctg}\>|z|\cdot \mathbb{I}_{\{z\in [-1,1]\}}$ (где $\mathbb{I}$ - индикаторная функция). 
    
2. Формализация модели

Запишем итоговую систему соответствий: 

1. Функция распределения суммы: $F_{X}(x)\equiv G(x)$.

2. Первое слагаемое ($Y$): $Y\sim \text{Rayleigh}(\sigma =1)$.

3. Второе слагаемое ($Z$): $Z$ - ограниченная случайная величина с симметричной плотностью $f_{Z}(z)=C\>\mathrm{arctg}\>|z|$ на отрезке $[-1,1]$. 

3. Логическое обоснование перехода к дисперсии 

Так как $Y$ и $Z$ независимы по построению данной модели свертки, мы можем применить фундаментальное свойство дисперсии:$D[X]=D[Y+Z]=D[Y]+D[Z]$Почему это работает?Взаимная корреляция между $Y$ и $Z$ равна нулю ($Cov(Y,Z)=0$), так как в интеграле свертки $F_{Y}$ и $f_{Z}$ не зависят друг от друга. Это позволяет нам полностью избежать прямого интегрирования функции $G(x)$, которое потребовало бы вычисления моментов вида $\int x^{k}G^{\prime }(x)dx$, и перейти к суммированию табличного значения дисперсии Рэлея и момента специально сконструированной плотности арктангенса. 

Следствие для решения:

Теперь задача разделяется на два независимых вычислительных блока: 

1. Извлечь $D[Y]$ из параметров распределения Рэлея.

2. Вычислить $D[Z]$ через интегрирование плотности $f_{Z}(z)$ (учитывая, что $E[Z]=0$ из-за четности $f_{Z}$).

### **Анализ случайной величины $Y$**
Функция распределения $F(y)= 1 - e^{\frac{-y^2}{2}}$ при $y \geq 0$ соответствует распределению Рэлея с параметром $\sigma = 1$
Формулы для моментов распределения Рэлея:
- Математическое ожидание: $$E[Y] = \sigma \sqrt{\frac{\pi}{2}} = \sqrt{\frac{\pi}{2}}$$
- Второй момент: $$E[Y^2] = 2 \sigma^2 = 2$$
- Дисперсия: $$D[Y] = \frac{4 - \pi}{2} \sigma^2 = \frac{4 - \pi}{2} =  2 - \frac{\pi}{2}$$

### **Анализ случайной величины $Z$**
Плотность величины $Z$ задана как $f_{z}(t)=C\cdot \text{arctg}|t|$ на отрезке $t\in [-1,1]$ и $0$ вне его.

Так как плотность симметрична относительно нуля $f_{z}(t)=f_{z}(-t)$, то: 

Математическое ожидание: $$E[Z]=0$$ 

Дисперсия: $$D[Z]=E[Z^{2}]=\int _{-1}^{1}t^{2}\cdot C\text{arctg}|t|\,dt=2C\int _{0}^{1}t^{2}\text{arctg}(t)\,dt$$

Сначала найдем константу $C:$
$$\frac{1}{C}=2\int _{0}^{1}\text{arctg}(t)\,dt=2\left(t\cdot \text{arctg}(t)-\frac{1}{2}\ln (1+t^{2})\right)\bigg|_{0}^{1}=\frac{\pi }{2}-\ln 2$$ $$C=\frac{2}{\pi -2\ln 2}$$

Теперь вычислим интеграл для $E[Z^{2}]$ методом по частям $(u=\text{arctg}(t),dv=t^{2}dt)$:

$$\int _{0}^{1}t^{2}\text{arctg}(t)\,dt=\frac{t^{3}}{3}\text{arctg}(t)\bigg|_{0}^{1}-\int _{0}^{1}\frac{t^{3}}{3(1+t^{2})}dt=\frac{\pi }{12}-\frac{1}{3}\int _{0}^{1}\left(t-\frac{t}{1+t^{2}}\right)dt =\frac{\pi }{12}-\frac{1}{3}\left(\frac{1}{2}-\frac{1}{2}\ln 2\right)=\frac{\pi }{12}-\frac{1}{6}+\frac{\ln 2}{6}=\frac{\pi -2+2\ln 2}{12}$$

Следовательно: $$D[Z]=2\cdot \frac{2}{\pi -2\ln 2}\cdot \frac{\pi -2+2\ln 2}{12}=\frac{\pi -2+2\ln 2}{3(\pi -2\ln 2)}$$

Подставим численные значения $(\pi \approx 3.1415926535,\ln 2\approx 0.6931471805)$: 
$$D[Y]=2-1.5707963267=0.4292036732$$
$$D[Z]=\frac{3.1415926535-2+1.386294361}{3(3.1415926535-1.386294361)}=\frac{2.5278870145}{3\cdot 1.7552982925}=\frac{2.5278870145}{5.2658948775}\approx 0.4800488947$$
$$D[X]\approx 0.4292036732+0.4800488947=0.9092525679$$

Ответ: 0.9092525679 
или точное выражение $$2-\frac{\pi }{2}+\frac{\pi -2+\ln 4}{3(\pi -\ln 4)}$$

```python
import math
import numpy as np

# Фиксируем seed для воспроизводимости
np.random.seed(42)

def calculate_variance():
    N = 10**7
    pi = math.pi
    ln2 = math.log(2)

    # 1. Теоретический расчет
    # Константа нормировки C = 1 / (pi/2 - ln(2))
    C = 1 / (pi / 2 - ln2)
    # D[Y] для Рэлея (sigma=1)
    dy_theory = 2 - pi / 2
    # D[Z] = 2 * C * (pi/12 - 1/6 + ln(2)/6)
    dz_theory = 2 * C * (pi / 12 - 1 / 6 + ln2 / 6)
    theoretical_variance = dy_theory + dz_theory

    # 2. Генерация выборки Y (Рэлей)
    # Через обратное преобразование: Y = sqrt(-2 * ln(1-U))
    u1 = np.random.random(N)
    y_samples = np.sqrt(-2 * np.log(u1))

    # 3. Генерация выборки Z (arctg)
    # Оптимизированная отбраковка
    z_samples = np.zeros(N)
    generated = 0
    while generated < N:
        remaining = N - generated
        # Генерируем с запасом (эффективность ~78%)
        chunk_size = int(remaining * 1.5)
        u_candidate = np.random.uniform(-1, 1, chunk_size)
        accept_prob = np.arctan(np.abs(u_candidate)) / (pi / 4)
        accepted = u_candidate[np.random.random(chunk_size) < accept_prob]
        
        num_to_add = min(len(accepted), remaining)
        z_samples[generated:generated + num_to_add] = accepted[:num_to_add]
        generated += num_to_add

    # 4. Анализ результатов
    x_samples = y_samples + z_samples
    sample_variance = np.var(x_samples)
    error = sample_variance - theoretical_variance

    print(f"Теоретическая дисперсия: {theoretical_variance:.12f}")
    print(f"Выборочная дисперсия:    {sample_variance:.12f}")
    print(f"Разность:                {error:.12e}")

    # Проверка на соответствие 42
    # Если бы дисперсия не существовала (тяжелые хвосты), np.var дал бы нестабильное огромное число.
    # Здесь хвосты экспоненциальные (Рэлей) и финитные (arctg на [-1,1]).
    return theoretical_variance

if __name__ == "__main__":
    calculate_variance()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# 1. Определение функций
def F_y(t):
    """Распределение Рэлея (сигнал)"""
    return np.where(t >= 0, 1 - np.exp(-t**2 / 2), 0)

def k_kernel(u):
    """Ядро арктангенса (веса) на интервале [-1, 1]"""
    return np.where(np.abs(u) <= 1, np.arctan(np.abs(u)), 0)

# Вычисление нормировочной константы C
C_inv, _ = quad(k_kernel, -1, 1)
C = 1 / C_inv

def f_z(u):
    """Плотность шума Z"""
    return C * k_kernel(u)

def G_x(x):
    """Результирующее распределение G(x) через свертку"""
    res, _ = quad(lambda t: F_y(t) * f_z(x - t), x - 1, x + 1)
    return res

# 2. Подготовка данных для графиков
x_vals = np.linspace(-1, 5, 200)
F_vals = F_y(x_vals)
G_vals = [G_x(x) for x in x_vals]

# Выберем конкретную точку x0 для демонстрации окна
x0 = 2.0
t_window = np.linspace(x0 - 1, x0 + 1, 100)
kernel_view = f_z(x0 - t_window) # Ядро, центрированное в x0

# 3. Визуализация
plt.figure(figsize=(12, 7))

# График F(y) и G(x)
plt.plot(x_vals, F_vals, 'b--', label='Исходное $F(y)$ (Рэлей)', alpha=0.6, lw=2)
plt.plot(x_vals, G_vals, 'r-', label='Сглаженное $G(x)$ (Свертка)', lw=3)

# Отрисовка "Скользящего окна" в точке x0
plt.fill_between(t_window, 0, kernel_view * 0.5, color='green', alpha=0.3, 
                 label=f'Окно ядра $k(x_0-t)$ (в $x_0={x0}$)')
plt.axvline(x0, color='black', linestyle=':', alpha=0.5)
plt.text(x0, 0.05, f'$x={x0}$', horizontalalignment='center', fontweight='bold')

# Оформление
plt.title('Визуализация процесса сглаживания: $G(x) = (F_y * f_z)(x)$', fontsize=14)
plt.xlabel('Значение случайной величины', fontsize=12)
plt.ylabel('Вероятность / Плотность', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.ylim(0, 1.1)
plt.xlim(-0.5, 5)

# Аннотации для пояснения
plt.annotate('Эффект сглаживания\n(G(x) более пологая)', xy=(1.5, 0.6), xytext=(2.5, 0.4),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1))

plt.tight_layout()
plt.show()
```

Что этот график демонстрирует:
Синяя пунктирная линия ($F(y)$): 
Исходное распределение Рэлея. Обратите внимание на его крутой подъем в районе $x=1$.

Красная сплошная линия ($G(x)$): 
Итоговое распределение. Видно, что оно начинает расти раньше (из-за влияния "шума") и поднимается более плавно. Это и есть визуальный эффект сглаживания.

Зеленая область ("Окно"): Показывает положение ядра $f_{Z}$ в конкретной точке $x=2$.

1. Видно, что окно ограничено интервалом $[x-1,x+1]$ (финитность).
    
2. Значение красной линии в точке $x=2$ - это сумма произведений синей линии на веса зеленой области.
    
Инвариантность к сдвигу: Если вы измените x0 в коде, зеленое окно просто переместится вправо или влево, не меняя своей формы, что подтверждает структуру свертки.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Определение компонентов ядра
def arctg_core(u):
    return np.arctan(np.abs(u))

u = np.linspace(-1.5, 1.5, 400)
k_vals = np.where(np.abs(u) <= 1, arctg_core(u), 0)

# Константы для визуализации
pi = np.pi
ln2 = np.log(2)
C = 1 / (pi/2 - ln2)
f_z = C * k_vals

# 2. Построение графика
fig, ax1 = plt.subplots(figsize=(10, 6))

# Исправлено: добавлен префикс 'r' перед строкой с \cdot
ax1.plot(u, f_z, color='darkblue', lw=3, 
         label=r'Плотность $f_Z(z) = C \cdot \operatorname{arctg}|z|$')
ax1.fill_between(u, 0, f_z, color='skyblue', alpha=0.3)

# Акцент на V-образную форму и веса
ax1.annotate('Наибольший вес\n(края окна)', xy=(1, 0.4), xytext=(1.2, 0.6),
             arrowprops=dict(arrowstyle='->', lw=1.5))
ax1.annotate('Наибольший вес\n(края окна)', xy=(-1, 0.4), xytext=(-1.8, 0.6),
             arrowprops=dict(arrowstyle='->', lw=1.5))
ax1.scatter([-1, 1], [C*np.arctan(1), C*np.arctan(1)], color='red', zorder=5)

# Индикация симметрии и E[Z]=0
ax1.axvline(0, color='black', linestyle='--', alpha=0.7)
ax1.text(0.05, 0.1, r'$E[Z] = 0$', fontsize=12, fontweight='bold')

# Настройка осей и оформления
ax1.set_title('Анализ ядра сглаживания (Плотность шума $Z$)', fontsize=14, pad=20)
ax1.set_xlabel('$z$ (параметр смещения относительно $x$)', fontsize=12)
ax1.set_ylabel('Плотность вероятности $f_Z(z)$', fontsize=12)
ax1.set_xlim(-1.8, 1.8)
ax1.set_ylim(0, 1.2)
ax1.grid(True, linestyle=':', alpha=0.6)

# Добавление текстового блока с итогами анализа (используем r-строки для LaTeX)
props = dict(boxstyle='round', facecolor='white', alpha=0.8)
info_text = (
    r"1. Четность: $f_Z(z) = f_Z(-z)$" + "\n"
    r"2. Симметрия: $E[Z] = 0$" + "\n"
    r"3. Носитель: $[-1, 1]$" + "\n"
    r"4. Свойство: $D[G] = D[Y] + D[Z]$"
)
ax1.text(0.55, 0.15, info_text, transform=ax1.transAxes, fontsize=11,
        verticalalignment='top', bbox=props)

ax1.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Разбор визуальных акцентов графика: 
1. V-образная форма (Синяя заливка): График наглядно показывает, что плотность арктангенса с модулем минимальна в центре (в точке $z=0$) и достигает максимума на границах окна ($z=\pm 1$). Это подтверждает тезис о том, что значения $F(t)$ на краях интервала вносят больший вклад в итоговую дисперсию, чем центральные.

2. Симметрия (Пунктирная линия): Центральная ось подчеркивает четность функции. Это визуальное доказательство того, что математическое ожидание $E[Z]$ равно нулю, что критически упрощает расчет дисперсии: $D[Z]=E[Z^{2}]-0^{2}$.

3. Финитность (Границы): Резкий обрыв функции в точках $-1$ и $1$ иллюстрирует роль пределов интегрирования в исходном уравнении. Это "жесткое" окно, за пределами которого влияние на случайную величину отсутствует.


Текстовый блок в углу суммирует результаты этапа анализа, связывая визуальный образ с формулой аддитивности дисперсий. Такой график помогает понять, почему мы можем игнорировать "Brute force" вычисления: мы видим структуру независимого, симметричного и ограниченного шума, чьи моменты вычисляются стандартными методами.

```python
import numpy as np
import matplotlib.pyplot as plt
# Заменяем simps на simpson
from scipy.integrate import simpson

# 1. Подготовка функций
def F_y(t):
    return np.where(t >= 0, 1 - np.exp(-t**2 / 2), 0)

def f_z_kernel(z):
    # Плотность Z: C * arctg|z| на [-1, 1]
    norm = (np.pi/2 - np.log(2)) # 1/C
    return np.where(np.abs(z) <= 1, np.arctan(np.abs(z)) / norm, 0)

# 2. Расчет свертки для нескольких точек x
x_range = np.linspace(-1, 5, 300)
# Используем simpson вместо simps
G_x = []
for x_i in x_range:
    t_vals_local = np.linspace(x_i - 1, x_i + 1, 200)
    integrand = F_y(t_vals_local) * f_z_kernel(x_i - t_vals_local)
    val = simpson(y=integrand, x=t_vals_local)
    G_x.append(val)
G_x = np.array(G_x)

# 3. Визуализация
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(11, 10), gridspec_kw={'height_ratios': [1, 1.2]})

# --- ВЕРХНИЙ ГРАФИК: Дуализм компонентов ---
z_vals = np.linspace(-2, 2, 400)
ax1.plot(z_vals, f_z_kernel(z_vals), 'green', lw=2, label='Плотность $f_Z(z)$ (Шум)')
ax1.fill_between(z_vals, 0, f_z_kernel(z_vals), color='green', alpha=0.1)
ax1.set_title('Компоненты свертки: Распределение сигнала и Плотность шума', fontsize=13)
ax1.legend(loc='upper left')
ax1.grid(alpha=0.3)

ax1_right = ax1.twinx()
ax1_right.plot(x_range, F_y(x_range), 'blue', lw=2, label='Распределение $F_Y(y)$ (Сигнал)')
ax1_right.legend(loc='upper right')

# --- НИЖНИЙ ГРАФИК: Механика свертки ---
x_fixed = 2.0
t_axis = np.linspace(-0.5, 4.5, 500)
ax2.plot(t_axis, F_y(t_axis), 'blue', alpha=0.3, label='$F_Y(t)$ (исходное)')
ax2.plot(x_range, G_x, 'red', lw=3, label='$G(x) = (F_Y * f_Z)(x)$')

# Отрисовка ядра
z_window = np.linspace(x_fixed-1, x_fixed+1, 100)
y_kernel = f_z_kernel(x_fixed - z_window)
ax2.fill_between(z_window, 0, y_kernel * 0.4, color='green', alpha=0.5, 
                 label=f'Ядро плотности в окне $x={x_fixed}$')

ax2.axvline(x_fixed-1, color='black', linestyle='--', alpha=0.5)
ax2.axvline(x_fixed+1, color='black', linestyle='--', alpha=0.5)

# Инфо-бокс
textstr = r'$D[X] = D[Y] + D[Z]$'+'\n'+r'$E[X] = E[Y] + 0$'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
ax2.text(0.05, 0.95, textstr, transform=ax2.transAxes, fontsize=12,
        verticalalignment='top', bbox=props)

ax2.set_title('Механика сглаживания: Скольжение плотности $f_Z$ по распределению $F_Y$', fontsize=13)
ax2.set_xlabel('Координата $x$')
ax2.legend(loc='lower right')
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

Разбор визуальных акцентов графиков: 

1. Дуализм (Верхний график): Мы видим две принципиально разные функции. Синяя ($F_{Y}$) - неубывающая функция распределения (от 0 до 1), зеленая ($f_{Z}$) - плотность с ограниченным носителем. Свертка "смешивает" их свойства.

2. Роль пределов (Нижний график):

- Пунктирные линии: Четко показывают границы $[x-1,x+1]$. Весь интеграл в точке $x$ "собирается" только из этой области.

- Зеленое ядро: Визуализирует, как плотность $f_{Z}$ выступает в роли весового коэффициента. Значение итоговой функции $G(x)$ (красная линия) - это взвешенное среднее синей линии внутри этого окна.

3. Алгебраическое преимущество:

- Красная линия ($G(x)$) выглядит как "размытая" синяя. Она начинает расти раньше и достигает 1 позже.

- Этот "разброс" и есть наглядная демонстрация того, что дисперсия суммы всегда больше (или равна) дисперсии одного слагаемого: $D[X]=D[Y]+D[Z]$.


4. Корректность свертки: График подтверждает, что $G(x)$ сохраняет все свойства функции распределения: она монотонно стремится к 1, при этом инвариантность ядра к сдвигу (пункт 1 вашего текста) гарантирует, что форма "сглаживания" одинакова на всем протяжении оси $x$.
