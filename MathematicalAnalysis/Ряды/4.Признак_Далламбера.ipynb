# 4. Признак Даламбера

# Признак Даламбера (признак отношения)

## Формулировка

Рассмотрим ряд с положительными членами:
$$
\sum_{n=1}^{\infty} u_n, \quad u_n > 0.
$$

Пусть существует конечный предел
$$
\lim_{n \to \infty} \frac{u_{n+1}}{u_n} = l.
$$

Тогда:

1. Если $l < 1$, то ряд сходится.
2. Если $l > 1$, то ряд расходится.
3. Если $l = 1$, признак не даёт ответа.

---

## Идея признака

Признак Даламбера сравнивает ряд с геометрической прогрессией.

Если
$$
\frac{u_{n+1}}{u_n} \approx q,
$$
то
$$
u_n \approx C \cdot q^n,
$$
то есть ряд ведёт себя как геометрический:
$$
\sum q^n.
$$

- при $q < 1$ — сходится  
- при $q > 1$ — расходится  

---

## Доказательство: случай $l < 1$

### Шаг 1. Выбор промежуточного числа

Так как $l < 1$, выберем число $q$ такое, что
$$
l < q < 1.
$$

---

### Шаг 2. Использование определения предела

По определению предела:
$$
\lim_{n \to \infty} \frac{u_{n+1}}{u_n} = l
$$
означает:

для любого $\varepsilon > 0$ существует $N$, такое что при $n \geq N$:
$$
\left| \frac{u_{n+1}}{u_n} - l \right| < \varepsilon.
$$

Возьмём
$$
\varepsilon = q - l > 0.
$$

Тогда при $n \geq N$:
$$
\left| \frac{u_{n+1}}{u_n} - l \right| < q - l.
$$

Отсюда:
$$
\frac{u_{n+1}}{u_n} - l < q - l
$$
$$
\frac{u_{n+1}}{u_n} < q.
$$

---

### Шаг 3. Оценка членов ряда

Для $n \geq N$:
$$
u_{n+1} < q u_n.
$$

Последовательно применяя:

$$
u_{N+1} < q u_N,
$$
$$
u_{N+2} < q u_{N+1} < q^2 u_N,
$$
$$
u_{N+3} < q u_{N+2} < q^3 u_N,
$$

и в общем виде:
$$
u_{N+k} < q^k u_N.
$$

---

### Шаг 4. Сравнение с геометрическим рядом

Рассмотрим хвост ряда:
$$
u_{N+1} + u_{N+2} + u_{N+3} + \dots
$$

Используя оценки:
$$
u_{N+1} < q u_N,\quad
u_{N+2} < q^2 u_N,\quad
u_{N+3} < q^3 u_N,\ \dots
$$

получаем:
$$
u_{N+1} + u_{N+2} + \dots < u_N(q + q^2 + q^3 + \dots).
$$

Правая часть — геометрический ряд:
$$
\sum_{k=1}^{\infty} q^k,
$$
который сходится, так как $q < 1$.

---

### Шаг 5. Признак сравнения

По признаку сравнения:

если
$$
0 \le u_n \le v_n,
$$
и ряд $\sum v_n$ сходится, то сходится и $\sum u_n$.

Значит хвост исходного ряда сходится.

---

### Шаг 6. Удаление конечного числа членов

По свойству рядов:

сходимость ряда не зависит от конечного числа первых членов.

Следовательно, сходится весь ряд:
$$
\sum_{n=1}^{\infty} u_n.
$$

---

## Доказательство: случай $l > 1$

Выберем число $q$, такое что:
$$
1 < q < l.
$$

По определению предела:
$$
\frac{u_{n+1}}{u_n} > q > 1 \quad \text{при } n \geq N.
$$

Тогда:
$$
u_{n+1} > u_n,
$$

то есть члены ряда не убывают и, более того, не стремятся к нулю.

Но необходимый признак сходимости требует:
$$
u_n \to 0.
$$

Противоречие ⇒ ряд расходится.

---

## Случай $l = 1$

Признак не даёт ответа.

Примеры:

- Гармонический ряд:
$$
\sum \frac{1}{n} \quad \text{расходится}
$$

- Ряд:
$$
\sum \frac{1}{n^2} \quad \text{сходится}
$$

Но в обоих случаях:
$$
\lim_{n \to \infty} \frac{u_{n+1}}{u_n} = 1.
$$

---

## Итоговая интерпретация

Признак Даламбера — это проверка:

> "Насколько ряд похож на геометрическую прогрессию?"

- если "как $q^n$ с $q<1$" → сходится  
- если "как $q^n$ с $q>1$" → расходится  
- если "как $1^n$" → недостаточно информации
