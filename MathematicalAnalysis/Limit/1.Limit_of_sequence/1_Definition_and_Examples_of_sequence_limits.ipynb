### 1. Definition and Examples of sequence limits

Strict definition: <br>
**1. A function $f: \N \to X$, whose domain is the set of natural numbers, is called a sequence.**

Breakdown: <br>
- $f$ is a function, a rule of correspondence 
- $\N$ - is the domain. This is the set of natural numbers {1, 2, 3, ...}. The key point: the argument of the function is the index of the element in the sequence.
- $\to$ is the mapping symbol
- $X$ - is the codomain (the set of values). This can be any set: real numbers, vectors, words, etc. 
- Term: an object satisfying this definition is called a "sequence in X" 
Summary: A sequence is a law according to which every natural number $n$ (the index) is assigned exactly one element $f(n)$ from the set $X$

2. Notation and Terminology <br>
The values $f(n)$ of the function $f$ are called the terms of the sequence. It is customary to denote them by a symbol of an element of the set into which the mapping goes, endowing the symbol with the corresponding index of the argument, $x_n := f(n)$

Breakdown:
- $f(n)$ is the value of the function, the "output element"
- $x_n$ is an alternative (more convenient) notation for $f(n)$
- The n-th term of the sequence is a synonym for $x_n$ or $f(n)$

3. The sequence itself is denoted by the symbol $\{x_n\}$, and is also written as $x_1, x_2, ..., x_n, ...$ and is called a sequence in $X$ or a sequence of elements of the set $X$ 
- $x_n$ is the notation for the entire sequence as a single object. The curly braces here denote a set, a family of elements
- $x_1, x_2, ..., x_n, ...$ is an equivalent, but more detailed, way of writing the same object. The ellipsis $...$ indicates the infinite nature of this ordered list. 
- When we say "the sequence $\{x_n\}$", we are referring to the entire function as a whole, not to any one of its individual elements. 

4. Specification of the Codomain
Henceforth, in the nearing paragraphs, we will consider only sequences $f \ : \ \N \to \R$ of real numbers. 

---

Well, we have established that we are working with a numerical sequence $\{x_n\}$, where each $x_n \in \R$ (a real number). Now we aim to formalize the intuitive notion: "The terms of the sequence $x_n$ become arbitarily close to some number $A$ as the index $n$ increases without bound.

**2. First (General Topological) Definition** <br>
**Definition 2. A number $A \in R$ is called the limit of a numerical sequence $\{x_n\}$ if for every neiborhood $V(A)$ of the point $A$, there exists an index $N$ (choosing depend on $V(A)$), such that all terms of the sequence whose indices are greater than $N$ are contained within the specified neighborhood of the point $A$.**

Detailed Breakdown: <br>
- $A \in R$ - The limit is a specific real number.
- "for every neighborhood V(A) of the point A" 
    - A neighborhood of the point $A$ is any interval on the real line containing the point $A$ within it. For example, (A-1, A+1) or (A-0.05, A+0.05). It can be large or arbitrarily small.
    - "every" is the key operative word. No matter what neighborhood (however small, microscopic) is proposed, the definition must hold. 
    - "there exists an index $N$ ...":
        - $N$ is a treshold index, a boundary, a moment of event occurrence. It is a natural number $N \in \N$
        - $N$ depends on the chosen neighborhood $V(A)$. The smaller (stricter) the neighborhood, the larger this treshold index $N$ typically must be. The notation $N(V(A))$ emphasizes this dependency. 
- "... such that all terms of the sequences whose indices are greater than $N$ are contained within the specified neighborhood $V(A)$"
    - Formally: $\forall n > N: x_n \in V(A)$ 
    - Meaning: Starting from some point (index $N+1$), the sequence does not stray outside the given neighborhood $V(A)$. All subsequent terms $(x_{N+1}, x_{N+2}, ..._)$ are guaranteed to be "near" $A$.

Geometric Interpretation: <br>
We place a "hoop" (neighborhood $V(A)$) around the point $A$. The definition states that no matter how narrow we make this hoop, there exists an index $N$ such that all points of the sequence, starting from that index, lie inside this hoop. The "tail" of the sequence entirely falls within the specified neighborhood.

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем фигуру с правильным LaTeX
fig = plt.figure(figsize=(15, 12))
fig.suptitle('Топологическое определение предела последовательности\n'
             r'$A = \lim_{n\to\infty} x_n \ \Leftrightarrow \ \forall V(A)\ \exists N: \forall n>N\ x_n \in V(A)$',
             fontsize=16, fontweight='bold', y=0.98)

# Параметры последовательности
A = 3  # Предел
n_max = 60  # Количество членов последовательности
n = np.arange(1, n_max + 1)

# Создаем последовательность с небольшими случайными колебаниями для наглядности
np.random.seed(42)
x_n = A + np.exp(-0.1 * n) * (2 + 0.5 * np.sin(n/2)) + 0.2 * np.random.randn(n_max)

# Создаем основную ось для последовательности
ax_main = plt.subplot2grid((3, 3), (0, 0), colspan=3, rowspan=2)
ax_main.set_xlabel('Номер члена последовательности (n)', fontsize=12)
ax_main.set_ylabel('Значение $x_n$', fontsize=12)
ax_main.set_title(r'Последовательность $\{x_n\}$ и её предел $A$', 
                  fontsize=14, fontweight='bold')

# Рисуем последовательность
scatter = ax_main.scatter(n, x_n, s=40, alpha=0.8, color='blue', 
                         edgecolor='darkblue', linewidth=1, zorder=4,
                         label='Члены последовательности $x_n$')

# Рисуем линию предела
ax_main.axhline(y=A, color='red', linestyle='--', linewidth=3, 
                label=f'Предел $A = {A}$', zorder=3)

# Добавляем три различных окрестности V(A)
neighborhoods = [
    {'radius': 1.5, 'color': 'green', 'alpha': 0.15, 'label': 'Большая окрестность $V_1(A)$'},
    {'radius': 0.8, 'color': 'orange', 'alpha': 0.2, 'label': 'Средняя окрестность $V_2(A)$'},
    {'radius': 0.3, 'color': 'purple', 'alpha': 0.25, 'label': 'Маленькая окрестность $V_3(A)$'}
]

# Соответствующие N для каждой окрестности
def find_N_for_neighborhood(radius):
    """Находит минимальный N, такой что ∀n>N: x_n ∈ (A-radius, A+radius)"""
    for N in range(1, n_max):
        if all(np.abs(x_n[N:] - A) < radius):
            return N
    return n_max

N_values = [find_N_for_neighborhood(nh['radius']) for nh in neighborhoods]

# Рисуем окрестности и соответствующие N
for i, (neighborhood, N) in enumerate(zip(neighborhoods, N_values)):
    radius = neighborhood['radius']
    color = neighborhood['color']
    alpha = neighborhood['alpha']
    
    # Рисуем окрестность как закрашенную область
    ax_main.axhspan(A - radius, A + radius, alpha=alpha, color=color, zorder=1)
    
    # Рисуем границы окрестности
    ax_main.axhline(y=A + radius, color=color, linestyle='-', alpha=0.5, linewidth=1)
    ax_main.axhline(y=A - radius, color=color, linestyle='-', alpha=0.5, linewidth=1)
    
    # Рисуем вертикальную линию для N
    ax_main.axvline(x=N, color=color, linestyle=':', linewidth=2, 
                    alpha=0.8, zorder=2)
    
    # Подписываем N
    ax_main.text(N, ax_main.get_ylim()[0] + 0.1, f'$N_{i+1}$', 
                fontsize=11, fontweight='bold', color=color,
                ha='center', va='bottom')
    
    # Подсвечиваем "хвост" последовательности (n > N)
    if N < n_max:
        ax_main.axvspan(N, n_max, alpha=0.08, color=color, zorder=0)
    
    # Подписываем радиус окрестности
    ax_main.text(n_max, A + radius, f'$r_{i+1}=${radius}', 
                fontsize=10, color=color, va='bottom', ha='right')
    ax_main.text(n_max, A - radius, f'$-r_{i+1}$', 
                fontsize=10, color=color, va='top', ha='right')

# Размечаем точки, попавшие в каждую окрестность
for i, (neighborhood, N) in enumerate(zip(neighborhoods, N_values)):
    radius = neighborhood['radius']
    color = neighborhood['color']
    
    # Точки в "хвосте" (n > N)
    tail_indices = n[N:]
    tail_values = x_n[N:]
    
    # Подсвечиваем точки, которые действительно в окрестности
    in_neighborhood = np.abs(tail_values - A) < radius
    ax_main.scatter(tail_indices[in_neighborhood], tail_values[in_neighborhood],
                   s=60, color=color, edgecolor='black', linewidth=1.5,
                   marker='o', zorder=5, alpha=0.9)

ax_main.grid(True, alpha=0.3, linestyle='--')
ax_main.legend(loc='upper right', fontsize=10)
ax_main.set_xlim(0, n_max + 2)
ax_main.set_ylim(A - 2.5, A + 2.5)

# Создаем три вспомогательных графика для детализации каждой окрестности
for i, (neighborhood, N) in enumerate(zip(neighborhoods, N_values)):
    ax_detail = plt.subplot2grid((3, 3), (2, i))  # Третья строка, i-я колонка
    
    radius = neighborhood['radius']
    color = neighborhood['color']
    label = neighborhood['label'].split('$')[1]
    
    ax_detail.set_title(f'{label}\n$r = {radius}, N_{i+1} = {N}$', 
                       fontsize=11, fontweight='bold', color=color)
    
    # Рисуем только "хвост" последовательности для этой окрестности
    tail_n = n[N:]
    tail_x = x_n[N:]
    
    # Точки в окрестности и вне её
    in_nhood = np.abs(tail_x - A) < radius
    out_nhood = ~in_nhood
    
    # Точки в окрестности
    ax_detail.scatter(tail_n[in_nhood], tail_x[in_nhood], 
                     s=40, color=color, edgecolor='black', linewidth=1,
                     zorder=3, label=f'$x_n \\in V_{i+1}(A)$')
    
    # Точки вне окрестности (их не должно быть по определению!)
    if np.any(out_nhood):
        ax_detail.scatter(tail_n[out_nhood], tail_x[out_nhood], 
                         s=40, color='red', edgecolor='darkred', linewidth=1.5,
                         marker='x', zorder=4, label='$x_n \\notin V_{i+1}(A)$')
    
    # Окрестность
    ax_detail.axhspan(A - radius, A + radius, alpha=0.2, color=color, zorder=1)
    ax_detail.axhline(y=A, color='red', linestyle='--', linewidth=2, zorder=2)
    
    # Границы окрестности
    ax_detail.axhline(y=A + radius, color=color, linestyle='-', linewidth=1.5, alpha=0.7)
    ax_detail.axhline(y=A - radius, color=color, linestyle='-', linewidth=1.5, alpha=0.7)
    
    # Вертикальная линия N
    ax_detail.axvline(x=N, color=color, linestyle=':', linewidth=2, alpha=0.8)
    
    ax_detail.set_xlabel('n')
    ax_detail.set_ylabel('$x_n$')
    ax_detail.grid(True, alpha=0.3)
    ax_detail.set_xlim(N - 5, min(N + 15, n_max))
    ax_detail.set_ylim(A - radius*1.5, A + radius*1.5)
    
    if i == 0:
        ax_detail.legend(loc='upper right', fontsize=9)

# Добавляем пояснительный текст
explanation_text = (
    'Топологическое определение предела:\n'
    '1. Берем ЛЮБУЮ окрестность V(A) точки A (зеленая, оранжевая, фиолетовая области)\n'
    '2. Для каждой окрестности существует свой порог N (вертикальные пунктирные линии)\n'
    '3. ВСЕ члены последовательности с номерами n > N лежат внутри V(A)\n'
    '4. Чем меньше окрестность, тем больше требуется N'
)

fig.text(0.5, 0.02, explanation_text,
         ha='center', fontsize=11, style='italic',
         bbox=dict(boxstyle='round,pad=0.8', facecolor='lightyellow', 
                  edgecolor='gold', alpha=0.9))

# Добавляем стрелки-указатели
ax_main.annotate('Окрестности разного размера', xy=(45, A+1.2), xytext=(35, A+2),
                arrowprops=dict(arrowstyle='->', color='green', lw=2),
                fontsize=10, color='green', fontweight='bold')

ax_main.annotate('Соответствующие N для каждой окрестности', 
                xy=(N_values[2], A-0.5), xytext=(20, A-2),
                arrowprops=dict(arrowstyle='->', color='purple', lw=2),
                fontsize=10, color='purple', fontweight='bold')

ax_main.annotate('"Хвост" последовательности\n(все x_n при n > N в окрестности)', 
                xy=(n_max-5, A+0.1), xytext=(n_max-15, A+1.5),
                arrowprops=dict(arrowstyle='->', color='blue', lw=2),
                fontsize=10, color='blue', fontweight='bold', ha='center')

plt.subplots_adjust(bottom=0.12, top=0.9, hspace=0.4)
plt.show()

# Дополнительная визуализация: принцип работы определения
fig2, (ax2, ax3) = plt.subplots(1, 2, figsize=(14, 6))

# Левый график: принцип "для любой окрестности"
ax2.set_title('Принцип "для любой окрестности V(A)"', fontsize=14, fontweight='bold')

# Показываем несколько окрестностей разного размера
radii = [2.0, 1.0, 0.5, 0.2, 0.1]
colors = plt.cm.viridis(np.linspace(0, 1, len(radii)))

for i, (radius, color) in enumerate(zip(radii, colors)):
    # Рисуем окрестность
    ax2.axhspan(A - radius, A + radius, alpha=0.7/(i+1), color=color, zorder=i)
    
    # Находим соответствующий N
    N_val = find_N_for_neighborhood(radius)
    
    # Рисуем стрелку от окрестности к N
    ax2.annotate('', xy=(N_val, A), xytext=(n_max, A + radius - 0.05),
                arrowprops=dict(arrowstyle='->', color=color, lw=1.5, alpha=0.7))
    
    ax2.text(n_max + 2, A + radius, f'V_{i+1}(A)', 
            fontsize=9, color=color, va='center')
    ax2.text(N_val, A - 0.3, f'N_{i+1}', 
            fontsize=9, color=color, ha='center', fontweight='bold')

ax2.scatter(n, x_n, s=20, alpha=0.6, color='black', zorder=len(radii)+1)
ax2.axhline(y=A, color='red', linestyle='--', linewidth=2)
ax2.set_xlabel('n')
ax2.set_ylabel('$x_n$')
ax2.grid(True, alpha=0.3)
ax2.set_xlim(0, n_max + 5)
ax2.set_ylim(A - 2.5, A + 2.5)

# Правый график: зависимость N от размера окрестности
ax3.set_title('Зависимость N от размера окрестности', fontsize=14, fontweight='bold')

# Вычисляем N для разных радиусов
radii_test = np.logspace(-0.5, -2, 30)  # от ~0.3 до 0.01
N_test = [find_N_for_neighborhood(r) for r in radii_test]

scatter = ax3.scatter(radii_test, N_test, c=radii_test, cmap='viridis_r', 
                     s=100, edgecolor='black', linewidth=1, zorder=3)

# Линия тренда (примерная)
if len(radii_test) > 1:
    z = np.polyfit(np.log(radii_test), N_test, 1)
    p = np.poly1d(z)
    radii_fit = np.linspace(min(radii_test), max(radii_test), 100)
    ax3.plot(radii_fit, p(np.log(radii_fit)), 'r--', linewidth=2, 
            label='Тренд: $N \\sim \\log(1/r)$', alpha=0.7)

ax3.set_xlabel('Радиус окрестности r', fontsize=12)
ax3.set_ylabel('Пороговый номер N', fontsize=12)
ax3.grid(True, alpha=0.3)
ax3.set_xscale('log')
ax3.invert_xaxis()  # Меньше радиус -> больше N
ax3.legend()

# Цветовая шкала
cbar = plt.colorbar(scatter, ax=ax3)
cbar.set_label('Радиус окрестности', rotation=270, labelpad=15)

plt.tight_layout()
plt.subplots_adjust(bottom=0.15)

# Добавляем общую подпись
fig2.text(0.5, 0.02, 
         'Чем меньше окрестность V(A) (меньше радиус r), тем больше требуется N.\n'
         'Но для ЛЮБОЙ окрестности такой N существует — в этом смысл предела.',
         ha='center', fontsize=11, style='italic',
         bbox=dict(boxstyle='round,pad=0.6', facecolor='lightblue', alpha=0.8))

plt.show()

# Простая версия для лучшей наглядности определения
print("=" * 80)
print("СУЩЕСТВЕННАЯ ЧАСТЬ ОПРЕДЕЛЕНИЯ:")
print("=" * 80)
print("1. Выбираем произвольную окрестность V(A) точки A (любого размера)")
print("2. Находим номер N (зависит от размера окрестности)")
print("3. Проверяем: все x_n при n > N должны лежать внутри V(A)")
print("4. Если это верно для ЛЮБОЙ окрестности V(A), то A — предел")
print("=" * 80)
```

**3. Second (Standard "Epsilon-Neighborhood") Definition** <br>
**A number $A \in \R$ is called the limit of a sequence $\{x_n\}$ if for every $\varepsilon > 0$ there exists an index $N$ such that for all $n > N$ we have $|x_n - A| < \varepsilon$**

Detailed Breakdown: <br>
- "for every $\varepsilon$ > 0
    - $\varepsilon$ (epsilon) is a a positive number characterizing the allowable error or degree of closeness.
    - Again, the key word is "every". We can demand precision of $\varepsilon = 0.1$, $\varepsilon = 0.001$, $\varepsilon = 10^{-100}$
- "there exists an index N ...":
    - Similarly, $N \in \N$ is a treshold index, now depending on the choosen $\varepsilon$ $(N(\varepsilon))$
- "such that for all n > N we have $|x_n - A| < \varepsilon$
    - $|x_n - A| is the distance between the sequence term $x_n$ and the limit $A$ on the real line.
    - The inequality $|x_n - A| < \varepsilon$ means that $x_n$ lies within the $\varepsilon$-neighborhood of point $A$, i.e., in the interval $(A - \varepsilon), (A + \varepsilon)$.
    - The statement: $\forall n > N: |x_n - A| < \varepsilon.


```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем фигуру с большим пространством снизу
fig, axes = plt.subplots(2, 2, figsize=(14, 11))
fig.suptitle('Визуализация определения предела последовательности', 
             fontsize=16, fontweight='bold', y=0.98)

# Параметры
A = 2  # Предел
n_max = 100  # Максимальное количество членов последовательности
n = np.arange(1, n_max + 1)

# Пример сходящейся последовательности: x_n = 2 + 1/n
x_n = A + 1/n

# Различные ε (ширины окрестностей) для демонстрации
epsilons = [1.0, 0.5, 0.2, 0.1]
titles = [
    'ε = 1.0 (широкая окрестность)',
    'ε = 0.5',
    'ε = 0.2',
    'ε = 0.1 (узкая окрестность)'
]

for idx, (ax, epsilon, title) in enumerate(zip(axes.flat, epsilons, titles)):
    # Вычисляем соответствующий N для данного ε
    N = int(np.ceil(1/epsilon))
    
    # Ограничиваем N разумными пределами для визуализации
    if N > n_max:
        N = n_max - 10
    
    # Рисуем последовательность
    ax.scatter(n, x_n, s=20, alpha=0.6, label=f'$x_n = 2 + 1/n$', zorder=3)
    
    # Рисуем линию предела
    ax.axhline(y=A, color='red', linestyle='--', linewidth=2, 
               label=f'Предел A = {A}', zorder=2)
    
    # Рисуем ε-окрестность (закрашенная область)
    ax.axhspan(A - epsilon, A + epsilon, alpha=0.2, color='green', 
               label=f'ε-окрестность: (A-ε, A+ε)', zorder=1)
    
    # Вертикальная линия, показывающая N
    ax.axvline(x=N, color='purple', linestyle=':', linewidth=2, 
               label=f'N = {N} (пороговый номер)', zorder=2)
    
    # Закрашиваем область "хвоста" последовательности (n > N)
    if N < n_max:
        ax.axvspan(N, n_max, alpha=0.1, color='orange', label='"Хвост" (n > N)', zorder=0)
    
    # Подписываем точки "хвоста", попавшие в окрестность
    tail_n = n[N:]  # n > N
    tail_x = x_n[N:]
    
    # Проверяем, какие точки действительно в окрестности
    in_neighborhood = np.abs(tail_x - A) < epsilon
    
    if np.any(in_neighborhood):
        ax.scatter(tail_n[in_neighborhood], tail_x[in_neighborhood], 
                  s=40, color='green', edgecolor='darkgreen', linewidth=1.5,
                  label='Точки в ε-окрестности', zorder=4)
    
    # Настройки графика
    ax.set_xlabel('Номер члена последовательности (n)', fontsize=10)
    ax.set_ylabel('Значение $x_n$', fontsize=10)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.set_xlim(0, n_max)
    
    # Динамический предел по y для лучшей видимости
    y_padding = max(epsilon * 1.5, 0.5)
    ax.set_ylim(A - y_padding * 2, A + y_padding * 2)
    
    # Легенда только для первого графика
    if idx == 0:
        ax.legend(loc='upper right', fontsize=9)

# Добавляем пояснительный текст ПОД графиками
fig.text(0.5, 0.02, 
         'Определение предела: ∀ε>0 ∃N: ∀n>N |xₙ - A| < ε\n'
         'Чем меньше ε (уже окрестность), тем больше требуется N',
         ha='center', fontsize=11, style='italic', 
         bbox=dict(boxstyle='round,pad=0.5', facecolor='wheat', alpha=0.5))

# Увеличиваем расстояние снизу для текста
plt.subplots_adjust(bottom=0.08, top=0.92, hspace=0.3, wspace=0.3)
plt.show()

# Второй график: зависимость N от ε
fig2, ax2 = plt.subplots(figsize=(10, 6))

epsilons_detailed = np.logspace(-1, -2, 20)
Ns_detailed = [int(np.ceil(1/eps)) for eps in epsilons_detailed]

ax2.plot(epsilons_detailed, Ns_detailed, 'b-o', linewidth=2, markersize=6)
ax2.set_xlabel('ε (ширина окрестности)', fontsize=12)
ax2.set_ylabel('N (пороговый номер)', fontsize=12)
ax2.set_title('Зависимость N от ε для последовательности $x_n = 2 + 1/n$', 
              fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_xscale('log')
ax2.invert_xaxis()

# Добавляем подпись внизу
fig2.text(0.5, 0.02, 
          'Наблюдение: Для последовательности xₙ = A + 1/n, N ≈ ⌈1/ε⌉',
          ha='center', fontsize=10, style='italic',
          bbox=dict(boxstyle='round,pad=0.5', facecolor='lightblue', alpha=0.5))

plt.subplots_adjust(bottom=0.12)
plt.tight_layout()
plt.show()

# Третий график: динамическое изменение "обруча"
fig3, ax3 = plt.subplots(figsize=(12, 7))

for i, epsilon in enumerate([1.0, 0.5, 0.2, 0.1]):
    N = int(np.ceil(1/epsilon))
    
    # Рисуем ε-окрестность
    alpha = 0.7 - i * 0.15
    ax3.axhspan(A - epsilon, A + epsilon, alpha=alpha, 
                color=plt.cm.viridis(i/4), 
                label=f'ε = {epsilon}, N = {N}')
    
    # Отмечаем N
    ax3.plot(N, A, 's', markersize=10, color=plt.cm.viridis(i/4), 
             markeredgecolor='black', zorder=5)

# Рисуем последовательность
ax3.scatter(n, x_n, s=30, color='black', alpha=0.7, 
           label='$x_n = 2 + 1/n$', zorder=4)
ax3.axhline(y=A, color='red', linestyle='--', linewidth=2, 
           label=f'Предел A = {A}', zorder=3)

# Вертикальные линии для N
for i, epsilon in enumerate([1.0, 0.5, 0.2, 0.1]):
    N = int(np.ceil(1/epsilon))
    ax3.axvline(x=N, color=plt.cm.viridis(i/4), linestyle=':', 
                alpha=0.5, zorder=2)

ax3.set_xlabel('Номер члена последовательности (n)', fontsize=12)
ax3.set_ylabel('Значение $x_n$', fontsize=12)
ax3.set_title('Геометрическая интерпретация: "Обручи" (ε-окрестности) вокруг предела', 
              fontsize=14, fontweight='bold')
ax3.grid(True, alpha=0.3)
ax3.legend(loc='upper right', fontsize=10)
ax3.set_xlim(0, 50)
ax3.set_ylim(1.5, 3.5)

# Добавляем подпись
fig3.text(0.5, 0.02,
          'Каждый "обруч" (ε-окрестность) ловит последовательность, начиная с соответствующего N',
          ha='center', fontsize=10, style='italic',
          bbox=dict(boxstyle='round,pad=0.5', facecolor='lightgreen', alpha=0.5))

plt.subplots_adjust(bottom=0.1)
plt.tight_layout()
plt.show()
```

#### Equivalence of limit definitions and its importance for ML

Proving the equivalence of limit definitions is essential for bridging geometric intuition and computational rigor. In the context of modern ML and AI, understanding this connection is critical: the neighborhood-based definition lays the groundwork for working in high-dimensional spaces (topology), while the $\varepsilon$-definition serves as the foundation for analyzing the convergence of gradient methods and estimating loss functions. <br>

V.A. Zorich emphasizes this transition to foster "mathematical flexibility"—the ability to shift from a qualitative description of proximity (vital for neural network architectures) to quantitative estimates of error bounds and algorithmic stability. The method is based on the metric properties of space, where the concept of object similarity (L-norms) is equivalent to their containment within a specific region (neighborhood). Let's prove it:
$$
\begin{gathered}
A = \lim \{x_n\} \iff \forall V(A) \; \exists N \in \N: \; \forall n > N \;(x_n \in V(A)) \\
\Updownarrow \\
A = \lim \{x_n\} \iff \forall \varepsilon > 0 \; \exists N \in \N: \; \forall n > N \;(|x_n - A| < \varepsilon)
\end{gathered}
$$

To prove that the equivalence of the two definitions of a limit, we will first clarify the fundamental concepts upon which they are base, and then provide a formal logic derivation, then we must demonstrate that rach implies the other $(P_1 \iff P_2)$

##### I. Definitions and preliminary analysis
1. Definition of neighborhood $V(A)$: <br>
According to Zorich, a neighborhood of a point $A \in \R$ is any open interval $(a, b)$ that contains $A$ (i.e., $a < A < b$).
$$\forall V(A)\ \exists N\in \mathbb{N}\ \forall n>N:(x_{n}\in V(A))$$

2. Definition of an $\varepsilon$ neighborhood: <br>
a symmetric neighbohood of the form $V_{\varepsilon} (A) = (A - \varepsilon, A + \varepsilon)$ is equivalent to the double inequality: $A - \varepsilon < x < A + \varepsilon$
$$\forall \varepsilon >0\ \exists N\in \mathbb{N}\ \forall n>N:(|x_{n}-A|<\varepsilon)$$

3. The interval-value equivalence:
By the definition of an open interval, the statement $x \in (A - \varepsilon, A + \varepsilon)$ is equivalent to double inequality $A - \varepsilon < x < A + \varepsilon$

Subtracing $A$ from all parts, we obtain $-\varepsilon < x - A < \varepsilon$. By the definition of the absolute value (modulus) as the distance on the real line, this is exactly: $|x - A| < \varepsilon$ 

Conclusion: Being inside an $\varepsilon$-neighborhood and satisfying the $\varepsilon$-inequality$ are the mathematically identical statements.

##### II. Formal logical proof

Step 1: $P_1 \to P_2$ (from neighborhoods to $\varepsilon$)

Logic: if the condition holds for any arbitrary neighborhood, it must hold for the specific case of a symmetic $\epsilon$-interval.
1. Assume $P_1$ is true: for every neighborhood $V(A)$, there exists a natural number $N$.
2. Let $\varepsilon > 0$ be given arbitary 
3. Construct a specific neighborhood $V_{\varepsilon}(A) = (A - \varepsilon, A + \varepsilon)$. Since this is an open interval containing $A$, It qualifies as a neighborhood by definition.
4. Apply $P_1$ to $V_{\varepsilon}(A): there exists $N \in N$ such that for all $n > N$, we have $x_n \in V_{\varepsilon} (A)$
5. As shown in the preliminary analysis, $x_n \in (A - \varepsilon, A + \varepsilon) is equivalent to $|x_n - A| < \varepsilon$ 
6. Thus, for any $\varepsilon > 0$, we have found an $N$ such that $|x_n - A| < \varepsilon$ for all $n > N$. This is $P_2$

Step 2: $P_2 \to P_1$ (From $\varepsilon$ to Neighborhoods)

Logic: Every general neighborhood $V(A)$ contains a smaller, symmetric $\varepsilon$-neighbohood "inside" it.
1. Assume $P_2$ is true: for every $\varepsilon > 0$, there exists a natural number $N$ 
2. Let $V(A)$ be an arbitrary neighborhood of $A$. By definition, $V(A)$ is an open interval $(a, b)$ suchc that $a < A < b$ 
3. We need to find an $\varepsilon$ such that the $\varepsilon$-neighborhood is entirely neighborhood contained witihn $(a, b)$ 
Set: 
$$\varepsilon = min(A - a, b - A)$$
(Note: $\varepsilon > 0$ because $A$ is strictly between $a$ and $b$)
4. By this construction, the interval $(A - \varepsilon, A + \varepsilon)$ is a subset of $(a, b)$, i.e., $V_{\varepsilon} (A) \subseteq V(A)$ 
5. Apply $P_2$ for this specific $\varepsilon$: there exists $N \in \N$ such that for all $n > N$, we have $|x_n - a| < \varepsilon$ 
6. This inequality implies $x_n \in \V_{\varepsilon} (A). Since $V_{\varepsilon} (A) \subseteq V(A)$, it follows that $x_n \in V(A)$
7. Thus, for any neighborhood $V(A)$, we have found an $N$ such that $x_n \in V(A)$ for all $n > N$. This is $P_1$.

##### III. Key Takeaways for Academic Examination 
- Geometric Connection: The proof relies on the fact that the absolute value $|x-A|$ measures the distance between two points.
- The Nesting Principle: The equivalence is possible because every neighborhood contains an $\varepsilon$-neighborhood, and every $\varepsilon$-neighborhood is itself a neighborhood.
- Analytical Rigor: The transition between $x \in (A-\varepsilon ,A+\varepsilon )$ and $|x-A|<\varepsilon$ is justified by the formal definition of the modulus.

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем простую наглядную визуализацию
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Параметры
A = 2  # Предел
n = np.arange(1, 41)
x_n = A + 3 * np.exp(-0.15 * n)  # Сходящаяся последовательность

# Левый график: ε-N определение
ax1.set_title('Определение 2: ε-N формулировка', fontsize=14, fontweight='bold')
epsilon = 0.5

# Рисуем ε-окрестность
ax1.axhspan(A - epsilon, A + epsilon, alpha=0.3, color='blue', 
           label=f'ε-окрестность: (A-ε, A+ε), ε={epsilon}')

# Находим N для этого ε
N_eps = np.where(np.abs(x_n - A) < epsilon)[0][0] + 1
ax1.axvline(x=N_eps, color='red', linestyle='--', linewidth=2,
           label=f'N = {N_eps}')

# Показываем точки в окрестности
for i in range(N_eps, len(n)):
    ax1.scatter(n[i], x_n[i], color='green', s=50, zorder=5,
               label='x_n: |x_n-A| < ε' if i == N_eps else "")

ax1.plot(n, x_n, 'o-', markersize=4, linewidth=1, color='black',
        label='Последовательность {x_n}')
ax1.axhline(y=A, color='darkred', linestyle='-', linewidth=2,
           label=f'Предел A = {A}')
ax1.set_xlabel('Номер n', fontsize=12)
ax1.set_ylabel('Значение x_n', fontsize=12)
ax1.legend(fontsize=9, loc='upper right')
ax1.grid(True, alpha=0.3)
ax1.text(0.5, 0.02, r'∀ε>0 ∃N: ∀n>N |x_n - A| < ε', 
         transform=ax1.transAxes, ha='center', fontsize=11,
         bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))

# Правый график: Топологическое определение
ax2.set_title('Определение 1: Топологическая формулировка', fontsize=14, fontweight='bold')

# Рисуем произвольную окрестность V(A)
V_left, V_right = A - 0.8, A + 0.6
ax2.axhspan(V_left, V_right, alpha=0.3, color='orange',
           label=f'Произвольная окрестность V(A)\n= ({V_left:.1f}, {V_right:.1f})')

# Находим ε-окрестность внутри V(A)
epsilon_inner = min(A - V_left, V_right - A)
ax2.axhspan(A - epsilon_inner, A + epsilon_inner, alpha=0.2, color='blue',
           label=f'ε-окрестность внутри V(A)\nε = {epsilon_inner:.2f}')

# Находим N для этой ε-окрестности
N_V = np.where(np.abs(x_n - A) < epsilon_inner)[0][0] + 1
ax2.axvline(x=N_V, color='red', linestyle='--', linewidth=2,
           label=f'N = {N_V}')

# Показываем, что точки после N в V(A)
for i in range(N_V, len(n)):
    ax2.scatter(n[i], x_n[i], color='green', s=50, zorder=5,
               label='x_n ∈ V(A)' if i == N_V else "")

ax2.plot(n, x_n, 'o-', markersize=4, linewidth=1, color='black')
ax2.axhline(y=A, color='darkred', linestyle='-', linewidth=2)
ax2.set_xlabel('Номер n', fontsize=12)
ax2.set_ylabel('Значение x_n', fontsize=12)
ax2.legend(fontsize=9, loc='upper right')
ax2.grid(True, alpha=0.3)
ax2.text(0.5, 0.02, r'∀V(A) ∃N: ∀n>N x_n ∈ V(A)', 
         transform=ax2.transAxes, ha='center', fontsize=11,
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

# Связь между графиками
ax1.annotate('Частный случай\nокрестности', xy=(N_eps, A), xytext=(10, A+1.5),
            arrowprops=dict(arrowstyle='->', color='blue', lw=1.5),
            fontsize=10, ha='center')

ax2.annotate('Содержит\nε-окрестность', xy=(N_V, A), xytext=(30, A-1.5),
            arrowprops=dict(arrowstyle='->', color='orange', lw=1.5),
            fontsize=10, ha='center')

plt.tight_layout()
plt.show()
```

**Final Emphasis** <br>
The latter formulation of the definition of a limit means that no matter what accuracy $\varepsilon > 0$ we prescribe, there exists an index $N$ such that the absolute error of approximating the number $A$ by the terms of the sequence $\{x_n\}$ is less than $\varepsilon$, whenever $n > N$<br>
**This is the essence of a limit**
- "No matter what accuracy we prescribe" - a claim of arbitrary closeness 
- "there exists an index $N$" - a guarantee that this closeness is eventually attained
- "the absolute error of ... is less than $\varepsilon$" - a quantitative measure of closeness 
- "...whenever $n > N$" - a condition ensuring that this is not a mere coincidence for a few isolated terms (such as $x_5$ and $x_10$ accidentally falling into the neighborhood), but a persistent and collective property of the entire "tail" of the sequence.

**Summary (Strictly according to the text):**

Definition 1 (in terms of neighborhoods): A limit A is a number such that any pre-specified "hoop" (V(A)) around it contains almost the entire sequence (all terms starting from some index N).

Definition 2 (in terms of $\varepsilon$-neighborhoods): A limit A is a number such that for any required precision $(\varepsilon > 0)$, almost all terms of the sequence lie within a distance less than $\varepsilon$ from A.

Equivalence: The definitions are equivalent because an $\varepsilon$-neighborhood is a particular, yet universal, type of neighborhood.

Meaning: A limit describes the stable and unbounded approximation of the sequence's terms to the number A.

---

**DEFINITION 3.** If $\lim_{x \to \infty} x_n = A$, then the sequence $\{x_n\}$ is said to converge to $A$ or tends to $A$, and written $x_n \to A$ as $n \to \infty$

A sequence has a limit is called a convergent sequence. A sequence that does not have a limit is called a divergent sequence.
