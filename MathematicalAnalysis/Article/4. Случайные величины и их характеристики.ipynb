**Пример 1.** В модели двукратного подбрасывания монеты с пространством исходов $\Omega = \{ГГ, ГР, РГ, РР\}$ определим случайную величину $\xi = \xi(\omega)$ с помощью таблицы. 

| $\omega$        | ГГ | ГР | РГ | РР |
|-----------------|----|----|----|----|
| $\xi (\omega)$ | 2  | 1  | 1  | 0  | 

Здесь $\xi(\omega)$ по своему смыслу есть не что иное, как число гербов отвечающее исходу $\omega$

Другим простейшим примером случайной величины $\xi$ является индикатор (характеристическая функция) некоторого множества $A \in \mathcal{A}:$ 
$$\xi = I_A(\omega)$$

где 

$$I_A (\omega) = 
\begin{cases}
    1, & \omega \in A \\
    0, & \omega \notin A.
\end{cases}$$

```python
A = {"ГГ", "ГР", "РГ"}

def indicator(omega):
    if omega in A:
        return 1
    else:
        return 0

for w in ["ГГ","ГР","РГ","РР"]:
    print(w, indicator(w))
```

## Случайная величина и её распределение

Когда вводится случайная величина $\xi$, нас интересует не само пространство исходов $\Omega$, а **какие значения принимает $\xi$ и с какими вероятностями**.

Пусть задана вероятностная модель

$$
(\Omega, \mathcal{A}, P)
$$

и случайная величина

$$
\xi : \Omega \to \mathbb{R}.
$$

Это означает, что каждому исходу $\omega$ сопоставляется число $\xi(\omega)$.

---

### Множество значений случайной величины

Обозначим множество всех значений случайной величины

$$
X = \{x_1, x_2, \dots, x_m\}.
$$

Например, если $\xi$ — число орлов при двух бросках монеты, то

$$
X = \{0,1,2\}.
$$

---

### Вероятности значений случайной величины

Для любого множества $B \subset X$ определим вероятность

$$
P_\xi(B) = P\{\omega : \xi(\omega) \in B\}.
$$

Это вероятность того, что случайная величина принимает значение из множества $B$.

---

### Распределение случайной величины

Достаточно знать вероятности

$$
P_\xi(x_i) = P\{\omega : \xi(\omega) = x_i\}.
$$

Набор чисел

$$
\{P_\xi(x_1), \dots, P_\xi(x_m)\}
$$

называется **распределением вероятностей случайной величины** $\xi$.

---

### Пример

Два броска монеты

$$
\Omega = \{ГГ, ГР, РГ, РР\}
$$

Случайная величина — число орлов

$$
\xi(ГГ)=2,\quad
\xi(ГР)=1,\quad
\xi(РГ)=1,\quad
\xi(РР)=0
$$

Множество значений

$$
X = \{0,1,2\}
$$

Распределение

$$
P_\xi(0)=\frac14,\quad
P_\xi(1)=\frac12,\quad
P_\xi(2)=\frac14
$$

---

### Ключевая идея

Мы переходим от вероятностей исходов

$$
P(\omega)
$$

к вероятностям значений случайной величины

$$
P(\xi = x)
$$

```python
# Пространство исходов
omega = ["ГГ", "ГР", "РГ", "РР"]

# Случайная величина \xi (число орлов)
def xi(w):
    return w.count("Г")

# Вероятности исходов P (\xi)
P = {
    "ГГ": 0.25,
    "ГР": 0.25,
    "РГ": 0.25,
    "РР": 0.25
}

# Распределение  случайной величины \xi 
from collections import defaultdict

P_xi = defaultdict(float)

for w in omega:
    x = xi(w)
    P_xi[x] += P[w]

print(dict(P_xi))
```

Случайная величина $\xi$, принимающая два значения 1 и 0с вероятностями "успеха" - p и "неуспеха" - q, называется бернуллиевской. 

$$P_{\xi}(x) = p^x q^{1-x}, \ \ x =0,1.$$

```python
import numpy as np

# PMF Bernoulli
def bernoulli_pmf(x, p):
    q = 1 - P
    if x not in (0, 1):
        return 0.0
    return (p**x) * (q**(1-x))

# Симуляция n испытаний Бернулли
def bernoulli_sample(n, p, rng=None):
    rng = np.random.default_rng() if rng is None else rng 
    # 1 = успех, 0 = не успех
    return rng.binomial(1, p, size=n)
```

Биномиальной (или биноминально распределенной) случайной величиной $\xi$ называется случайная величина, принимающая $n+1$ значение $0, 1, ..., n$ с вероятностями. $$P_{\xi}(x) = C_n^x p^x q^{n-x}, \ \ x = 0, 1, ..., n.$$

Интуитивно: какова вероятность получить $x$ успехов в $n$ независимых испытаниях 

Типичные примеры:

число орлов при $n$ бросках монеты

число бракованных деталей в партии

число кликов из $n$ показов рекламы.
