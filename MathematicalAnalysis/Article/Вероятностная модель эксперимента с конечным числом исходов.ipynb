# Вероятностная модель эксперимента с конечным числом исходов 

1.Рассмотрим некоторый эксперимент, при "комплексе условий" результаты которого описываются конечным числом различных исходов (явлений) $\omega_1, ..., \omega_N$. Здесь важно, что число исходов $N$ конечно, и для нас несущественна реальная природа этих исходов.

Исходы $\omega_1, ..., \omega_N$ называются элементарными событиями, а их совокупность (конечным) пространством элементарных событий или пространством исходов (универсум) $$\Omega = \{\omega_1, ..., \omega_N\}$$ 

Выделение пространства элементарных событий представляет собой первый шаг в формулировании понятия вероятностной модели какого либо эксперимента. 

**П1:** При однократном подбрасывании монеты протсранство исходов $\Omega$ состоит из двух элементарных событий:
$$\Omega = \{Г, Р\}$$



```python
import numpy as np

# Пространство исходов \Omega
omega = np.array(["Г", "Р"])

# Вероятность p элементарных событий  \omega
p = np.array([0.5 , 0.5])
```

**П2:** При n-кратном подбрасывании монеты пространство элементарных событий 
$$\Omega = \{\omega: \omega = (a_1, ..., a_n), a_i = \text{Г или Р}\}$$
Здесь общее число $N(\Omega)$ исходов равно $2^n$ 

```python
import itertools 

# например 3 раза подбрасываем
n = 3

omega = list(itertools.product(["Г", "Р"], repeat=n))

print(omega)
print(len(omega))  # 2^n = 2^3 = 8
```

**П3:** Пусть сначала подбрасывается монета. Если выпадает герб, то бросается шестигранная кость (с цифрами 1 - 6), если же выпадает решетка, то снова подбрасывается монета. Пространство элементарных событий данного эксперимента будет таким:
$$\Omega = \{\text{Г1, Г2, Г3, Г4, Г5, Г6, РГ, РР}\}$$

```python
omega = []

# Если Г
for i in range(1, 7):
    omega.append(("Г", i))

# Если Р
omega.append(("Р","Г"))
omega.append(("Р","Р"))

print(omega)
```

Это точная реализация пространства $\Omega$ из примера.

Когда ты программируешь такие задачи, ты фактически делаешь:

генерацию пространства элементарных событий.

Это главный объект теории вероятностей - $\Omega$

После этого можно:

- задавать события $A \in \Omega$

- вычислять вероятности

- делать симуляции.

```python
# Генерация пространства элементарных событий

import itertools

def coin_space(n):
    return list(itertools.product(["Г", "Р"], repeat=n))

print(coin_space(4))
```

Это создает $|\Omega| = 2^n$

**Замечание.** Комплекс условий — это совокупность предположений о том, как проводится эксперимент и какие исходы возможны.

В теории вероятностей он обычно не выписывается явно, а предполагается заданным.

Важно: один и тот же эксперимент при разных комплексах условий может приводить к разным вероятностным моделям.

Например, при подбрасывании монеты:

если монета считается симметричной,$$P(Г) = P(Р) = \frac{1}{2}$$

Если монета несблансирована $$P(Р) \neq P(Р)$$

Следовательно, вероятность определяется не только событием, но и условия эксперимента

2. Рассмотрим примеры связанные с разными способами выбора $n$ шаров из урны, содержащей $M$ различных шаров. 

**П4.** Выбор с возвращением - эксперимент в котором на каждом шаге извлеченный шар возвращается обратно. 

В этом случае каждая выборка из $n$ шаров может быть записана в виде $(a_1, ..., a_n)$, где $a_i$ - номер шара, извлеченного на $i-$м шаге. Она в случае выбора с возвращением каждое $a_i$ может принимать любое из $M$ значений $1, 2, ..., M$. Описание пространства элементарных событий зависит от того, считаем ли мы выборки тождественного состава, такие как, скажем (4, 1, 2, 1) и (1, 4, 2, 1), различными или одинаковыми. В связи с этим принято различать два случая:

1. В случае выборки, состоящие из одних и тех же элементов, но отличающиеся порядком следования этих элементов, объявляются различными. 

2. Во втором случае порядок следования элементов не принимается во внимание и такие вбыорки объявляются тождественными. Чтобы подчеркнуть, какие конкретно выборки мы рассматриваем, будем для упорядоченных выборок использовать обозначение $(a_1, ..., a_n)$, а для неупорядоченных - $[a_1, ..., a_n]$. 

Итак, в случае упорядоченных выборок с возвращением пространство элементарных событий $\Omega$ имеет следующую структуру:

$$\Omega = \{\omega: \omega = (a_1, ..., a_n), a_i = 1, ..., M\}$$
и число различных исходов (выборок), называемых в комбинаторике размещениями из $M$ по $n$ с повторениями, равно $$N(\Omega) = M^n$$ 



```python
import itertools 

M = 2 
n = 2

omega = list(itertools.product(range(1, M + 1), repeat=n))

print(omega)
print("число исходов:", len(omega))

# Проверка формулы M^n
print(len(omega), "=", M**n)
```

Если же рассматриваются неупорядоченные выборки с возвращением (в комбинаторике - сочетания из $M$ по $n$ с повторениями), то $$\Omega = \{\omega: \omega = [a_1, ..., a_n], a_i = 1, ..., M\}$$

$$|\Omega| = \binom{M+n-1}{n}$$

```python
import itertools

M = 3 
n = 2 

omega = list(itertools.combinations_with_replacement(range(1, M+1), n))

print(omega)
print("число исходов:", len(omega))
```

```python
import math

# Проверка
formula = math.comb(M+n-1, n)

print(len(omega), "=", formula)
```

Понятно, что число $N(\Omega)$ (различных) неупорядоченых выборок меньше, чем число упорядоченных. ПОкажем, что для этого случая $$N(\Omega) = C^n_{M+m - 1}$$

где $C_k^{l} \equiv \frac{k!}{l!(k-l)!}$ - число сочетаний из $k$ элементов по $l$. 

Будем вести доказательство по индукции. Обозначим $N(M, n)$ число интересующих нас исходов. 

```python
# Проверка эксперимента 

import numpy as np

M = 5
n = 4

rng = np.random.default_rng()

sample = rng.integers(1, M+1, size=n)

print(sample)
```

Будем вести док-во по индукции. Обозначим $N(M, n)$ число интересующих нас исходов. Ясно, что для всех $k \leq M$

$$N(k, 1) = k = C^1_k$$

**Определение.** Вероятностное пространство это - $$(\Omega, \mathcal{A}, P)$$ 

где
$\Omega = \{ \omega_1, ..., \omega_N\}$, 
$\mathcal{A} - $ некоторая алгебра подмножеств $\Omega$ и $P = \{P(A); A \in \mathcal{A}\}$, определяет вероятностную модель (элементарных событий) $\Omega$ и алгеброй событий $\mathcal{A}$. Здесь $P(\{\omega_i\}) = p(\omega_i), i = 1, ..., N$
