## 1. Функции нескольких переменных

**Определение 1** Если каждой паре $(x, y)$ значений двух независимых друг от друга переменных величин $x$ и $y$ из некоторой области их изменения $D$, соответствует определенное значение величины $z$, то мы говорим, что $z$ есть функциия двух независимых переменных $x$ и $y$, определенная в области $D$. 

$$z = f(x, y), \quad z = F(x,y) \quad \text{и т.д.}$$

---

**Определение 2** Совокупность пар $(x, y)$ значений $x$ и $y$, при которых определяется функция $z = f(x,y)$, называется областью определения или областью существования этой функции.

Геометрически:

1) Если каждую пару значений $x$ и $y$ мы будем изображать точкой $M(x,y)$ в плоскости $Oxy$, то область определения функции изобразится в виде некоторой совокупности точек на плоскости

2) Эту совокупность точек будем называть областью определения функции. В частности, областью определения может быть и вся плоскость.

3) Линию, ограничивающую данную область, будем называть границей области. 

4) Точки области, не лежащие на границе, будем называть внутренними точками области

5) Область, состоящая из одних внутренних точек, называется открытой или незамкнутой.

6) Если же к области относятся и точки границы, то область называется замкнутой 

7) Область называется огрниченной, если существует такая постоянная $C$, что расстояние любой точки $M$ области от начала координат $O$ меньше $C$, т.е. $|OM| < C$.

---

Определение функции двух переменных легко обобщить на случай трех или более переменных

**Определение 3**. Если каждой рассматриваемой совокупности значений переменных $x, y, z, ..., u, t$ соответствует определенное значение переменной $\omega$, то будем называть $\omega$ функцией независимых переменных $x, y, z, ..., u ,t$ и писать $\omega = F(x, y, z, ..., u, t)$ или $\omega = f(x, y, z, ..., u, t)$ и т.п.

Та же как идля функции трех, четырех и более переменных областью определения является некоторая совокупность троек чисел $(x,y,z)$. Заметим тут же, что каждая тройка чисел задает некоторую точку $M(x,y,z)$ в протсранстве $Oxyz$. Следовательно, областью определения функции трех переменных является неторая совокупность точек пространства.

В каждой точке $(x,y)$ восставим перпендикуляр к плоскости $Oxy$ и на нем отложим отрезо, равный $f(x, y)$.

Тогда мы получим в пространстве точку $P$ с координатами $$x, y, z = f(x,y)$$

Геометрическое место точек $P$, координаты которых удовлетворяют уравнению (1), называется графиком функции двух переменных. Из курса аналитической геометрии известно, что уравнение (1) в пространстве определяет некоторую поверхность. Таким образом,, графиком функции двух переменных является поверхность, проетирующаяся на плоскость $Oxy$ в область определения функции. Каждый перпендикуляр к плоскости $Oxy$ пересекает поверхность $z = f(x,y)$ не более чем в одной точке.

Например: $z = x^2 + y^2$ - парабалоид вращения

# 3. Частное и полное приращение функции 



Рассмотрим линию $PS$ пересечения поверхности 

$$z = f(x,y)$$

с плоскостью $y= const$, параллельной плоскости $Oxz$ см. рис.

<div align="center">
  <img src="169.png" alt="alt text" width="250">
</div>

Так как в этой плоскости $y$ сохраняет постоянное значение, то $z$ вдоль кривой $PS$ будет меняться только в зависимости от изменения $x$. Дадим независимой переменной $x$ приращение $\Delta x;$  тогда $z$ получит приращение, которое называют частным приращением $z$ по $x$ и обозначают через $\Delta_xz$ (на рисунке отрезо SS'), так что 

$$\Delta_xz = f(x+\Delta x, y) - f(x, y) \tag{1}$$

Аналогично, если $x$ сохраняет постоянное значение, а $y$ полученым приращением $\Delta y$, то $z$ получает приращение, называемое частным приращением $z$ по $y$. Это приращение обозначают символом $\Delta_yz (на рисунке отрезок TT'):

$$\Delta_yz = f(x, y + \Delta y) - f(x, y) \tag{2}$$

Приращение $\Delta_yz$ функция получает "вдаоль линии" пересечения поверхности $z = f(x, y)$ с плоскостью $x = const$ , параллельной плоскости $Oyz$

Наконе, сообщив аргументу $x$ приращение $\Delta x$, а аргументу $y -$ приращение $\Delta y$, получим для $z$ новое приращение $\Delta z$, которое называется полным приращением функции $z$ и определяется формулой 

$$\Delta z = f(x + \Delta x, y + \Delta y) - f(x,y) \tag{3}$$

На рисунке $\Delta z$ изображается отрезком $QQ'$ 

Надо заметить, что, вообще говоря, полное приращение не равно сумме частных приращений, т.е. $\Delta z \neq \Delta_xz + \Delta_yz$

Пример. $z = xy$

$$\Delta_x z = (x+ \Delta x) y -xy = y \Delta x$$
$$\Delta_y z = x(y+ \Delta y) -xy = x \Delta y$$
$$\Delta z = (x + \Delta x)(y + \Delta y) -xy = y \Delta x + x \Delta y+ \Delta x \Delta y$$

При $x=1, y=2, \Delta x=0.2, \Delta y = 0.3$ имеем $\Delta_xz=0.4, \Delta_yz=0.4, \Delta_y z = 0.3, \Delta z =0.76$

Аналогичным образом определяются частные и полное приращения функции любого числа переменных. Так, Для функции трех переменных $u = f(x,y,t)$ имеем

$$\Delta_xu = f(x+\Delta x, y, t) - f(x,y,t),$$
$$\Delta_yu = f(x, y + \Delta y, t) - f(x,y,t),$$
$$\Delta_tu = f(x,y,t+\Delta t) - f(x,y,t),$$
$$\Delta u = f(x + \Delta x, y + \Delta y, t + \Delta t) - f(x,y,t)$$

# 4. Непрерывность функции нескольких переменных

**Окрестностью радиуса $r$ точки $M_0(x_0,y_0)$** называется совокупность всех точек $(x,y)$ удовлетворящих неравенству $$\sqrt{(x-x_o)^2+(y-y_0)^2} < r,$$ т.е. совокупность всех точек, лежащих внутри круга радиуса $r$ с центром в точке $M_o(x_0,y_0)$.

Если функция обладает каким-либо свойством "вблизи точки $(x_0, y_0)$" или "в орестности точки $(x_0, y_0)$", это означает, что найдется такой круг с центром $(x_0, y_0)$, во всех точках которого данная функция обладает указанным свойством.

**Определение 1.** Число $A$ называется пределом функции $f(x,y)$ при стремлении точки $M(x,y)$  точке $M_0(x_0, y_0)$, если для каждого числа $\varepsilon > 0$ найдется такое число $r > 0$. что для всех точек $M(x,y)$, для которых выполняется неравенство $\overline{MM_0} < r$, имеет место неравенство

Если число $A$ является пределом функции $f(x,y)$ при $M(x,y) \to M_0(x_0,y_0)$, то пишут 

$$\lim_{\substack{x \to x_0 \\ y \to y_0}} f(x, y) = A$$

**Определение 2.** Пусть точка $M_0(x_0, y_0)$ принадлежит области определения функции $f(x, y)$. Функция $z = f(x,y)$ называется непрерывной в точке $M_0(x_0, y_0)$, Если имеет место неравенство 

$$\lim_{\substack{x \to x_0 \\ y \to y_0}} f(x,y) = f(x_0,y_0), \tag{1}$$

Причем точка $M(x,y)$ стремится к точке $M_0(x_0,y_0)$ произвольным образом, оставаясь в области определения функции.

Если обозначим $x = x_0 + \Delta x, y = y_0 + \Delta y$, то равенство можно переписать так:

$$\lim_{\substack{\Delta x \to 0 \\ \Delta y \to 0}} f(x_0 + \Delta x, y_0 + \Delta y) = f(x_0, y_0) \tag{1'}$$

или

$$\lim_{\substack{\Delta x \to 0 \\ \Delta y \to 0}} [ f (x_0 + \Delta x, y_0 + \Delta y) - f(x_0, y_0)] = 0$$

$$\lim_{\Delta \rho \to 0} \Delta z = 0$$

Функция, непрерывная в каждой точке некоторой области, называется непрерывной в области.

# 5.Частные производные функции несольких переменных

**Опредеение.** Частной производной по $x$ от функции $z = f(x, y)$ называется предел отношения частного приращения $\Delta_X z$ к приращению $\Delta x$ при стремлении $\Delta x$ к нулю

Обозначения:
$$z'_x, f'_x (x,y), \frac{\partial z}{\partial x}, \frac{\partial f}{\partial x}$$

Таим образом, по определению:
$$\frac{\partial z}{\partial x} = \lim_{\Delta x \to 0} \frac{\Delta_x z}{\Delta x} = \lim_{\Delta x \to 0} \frac{f(x+\Delta x, y) - f(x,y)}{\Delta x}$$

по $y$ частная производная:

$$\frac{\partial z}{\partial y} = \lim_{\Delta y \to 0} \frac{\Delta_y z}{\Delta y} = \lim_{\Delta y \to 0} \frac{f(x, y + \Delta y) - f(x,y)}{\Delta y}$$

Заметив, что Axz вычисляется при неизменном у, a A9z принеизменном дг, мы можем определения частных производных сфор-сформулировать так: частной производной по хот функции z = f(x, у)называется производная по ху вычисленная в предположении, чтоу — постоянная. Частной производной по у от функции z = f(x,y)называется производная по у, вычисленная в предположении,что х — постоянная.

# 7. Полное приращение и полный дифференциал 

По определению полного приращения функции $z = f(x, y)$ имеем 

$$\Delta z = f(x + \Delta x, y + \Delta y) - f(x ,y)$$

Предположим, что $f(x, y)$ в рассматриваемой точке $(x, y)$ имеет непрерывные частные производные.

Выразим $\Delta z$ через частные производные. Для этого в правой части равенства (1) прибавим и вычтем $f(x, y + \Delta y)$:
