### 2.1 Предел функции на бесконенчности

Число $A$ называется пределом функции $f(x)$ при $x \to \infty$, если для любого сколь угодно малого положительного $\varepsilon > 0$ найдется такое положительное число $M > 0$ (зависящее от $\varepsilon$), что для всеех $x$, удовлетворяющих $|x| > M$ выполняется неравенство: $$|f(x) - A| < \varepsilon$

Аналогично для $x \to \infty$

$$\lim_{x \to -\infty} f(x) = A$$

если для любого $\varepsilon > 0$ существует число $M < 0$, что для всех $M$ выполняется: 

$$ |f(x) - A| < \varepsilon$$

Формально: $$ \lim_{x \to +\infty} f(x) = A \Longleftrightarrow \forall \varepsilon > 0 \ \exists M > 0 : x > P \Rightarrow |f(x) - A| < \varepsilon$$

### Применение предела функции на бесконечности в ML 

#### Сходимость эмпирического риска при обучении модели.

Эмпирический риск - средняя ошибка модели на обучающей выборке, используемая как приближение истинной (ожидаемой) ошибки на реальных данных 

<i>(Насколько хорошо модель работает на тех данных, которые мы уже видели? т.е. оценка, полученная по конечной выборке)</i>

Пусть:
- $\mathcal{X}$ - пространство признаков 
- $\mathcal{Y}$ - пространство ответов
- $\ell(y, \hat{y})$ - функция потерь
- $\{(x_i, y_i)\}^n_{i=1}$ - обучающая выборка 

Тогда эмпирический риск модели $f$ определяется как функция  
$$\mathcal{R}_n(f) = \frac{1}{n} \sum^n_{i=1} \ell (y_i, f(x_i))$$
